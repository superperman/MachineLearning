```python
year=np.arange(2009,2020)
month=['01','02','03','04','05','06','07','08','09','10','11','12']
day=28

date=[]
for y in year:
    for m in month:
        date.append(str(y)+'-'+str(m)+'-'+str(day))
#del date[-6:]
date=np.array(date)
date
```

```python
TR=pd.DataFrame(index=idx,columns=date[1:])
TR.head()
```

```python
for j in range(1,len(date)-1):
    R=[]
    for i in range(len(idx)):
        data=get_price(idx[i], start_date=str(date[j]), end_date=str(date[j+1]), frequency='daily', fields=['close'], skip_paused=False, fq='pre')
        R.append((data.iloc[-1]-data.iloc[0])/data.iloc[0]-MR[j-1]) #MR for Market Return at month[j]
    R=np.array(R)
    TR[str(date[j])]=R
```
