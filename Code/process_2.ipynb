```python
import numpy as np
import pandas as pd 
```

```python
n = 3
year=np.arange(2009,2020)  # change according to the year assigned
month=['01','02','03','04','05','06','07','08','09','10','11','12']
day=28
date=[]
for y in year:
    for m in month:
        date.append(str(y)+'-'+str(m)+'-'+str(day))
date=np.array(date) 
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
def filter_extreme(series,n):
    mean = series.mean(skipna=True)
    sd = series.std(skipna=True)
    max_range = mean + n*sd
    min_range = mean - n*sd
    return np.clip(series,min_range,max_range)
```

```python
for i in range(0, 124):
    data_name =date[i] + ".xlsx"
    data = pd.read_excel('.//data/' + data_name, index_col=0)
    print(data_name)
    data.loc[data['label']==0,'beta':'MFI14']=np.nan 
    data.drop(["ATR6","MAWVAD"],axis=1,inplace=True)
    for j in range(68):
        data.iloc[:,j+3] = filter_extreme(data.iloc[:, j+3] ,n)
        data.iloc[:,j+3].fillna(data.iloc[:,j+3].mean(),inplace=True)
#     try:
        data.iloc[:,3:] = min_max_scaler.fit_transform(data.iloc[:,3:])
#     except ValueError or TypeErrror:
#         pass
    data.loc[data['label']==0,'beta':'MFI14']=np.nan
    i_name = str(i+1) + '.csv'
    data.to_csv(i_name)
```
